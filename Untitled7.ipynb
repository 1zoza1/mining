```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Convert back to DataFrame
data_scaled = pd.DataFrame(scaled_data, columns=data.columns)

# Split features and target (example: predicting Satisfaction_Score)
x_train = data_scaled.drop('Satisfaction_Score', axis=1)
y_train = data_scaled[['Satisfaction_Score']]
```

```python
from sklearn.model_selection import train_test_split

# Split into training and testing sets (70% train, 30% test)
x_train, x_test, y_train, y_test = train_test_split(
    x_train, y_train, test_size=0.3, random_state=0
)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model_mlr = model.fit(x_train, y_train)
```

```python
y_pred1 = model_mlr.predict(x_test)
```

MLR Model

```python
from sklearn.metrics import mean_squared_error

mse1 = mean_squared_error(y_test, y_pred1)
print('The mean square error for Multiple Linear Regression:', mse1)
```

```python
from sklearn.metrics import mean_absolute_error

mae1 = mean_absolute_error(y_test, y_pred1)
print('The mean absolute error for Multiple Linear Regression:', mae1)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and fit the model
rf = RandomForestRegressor(random_state=0)
model_rf = rf.fit(x_train, y_train)
```

```python
y_pred2 = model_rf.predict(x_test)
```

```python
mse2 = mean_squared_error(y_test, y_pred2)
print('The mean square error of price and predicted value is: ', mse2)
```

```python
from sklearn.metrics import mean_absolute_error

mae2 = mean_absolute_error(y_test, y_pred2)
print('The mean absolute error of price and predicted value is:', mae2)
```

```python
scores = [('MLR', mae1),
          ('Random Forest', mae2)
           ]
```

```python
import pandas as pd

mae = pd.DataFrame(data=scores, columns=['Model', 'MAE Score'])
mae
```

```python
# Sort by MAE Score (descending)
mae.sort_values(by='MAE Score', ascending=False, inplace=True)

# Plot bar chart
import matplotlib.pyplot as plt
import seaborn as sns

f, axe = plt.subplots(1, 1, figsize=(10, 7))
sns.barplot(x=mae['Model'], y=mae['MAE Score'], ax=axe)

axe.set_xlabel('Model', size=20)
axe.set_ylabel('Mean Absolute Error', size=20)
axe.set_title('Model Comparison by MAE', size=22)

plt.grid(True)
plt.tight_layout()
plt.show()
```
